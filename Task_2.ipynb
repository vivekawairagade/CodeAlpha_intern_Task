Load Dataset

Basic Exploration

```python
# First rows
print(df.head())

# Info about columns
print(df.info())

# Summary statistics
print(df.describe())

# Column names
print(df.columns)
```

Check Data Quality

```python
# Missing values
print(df.isnull().sum())

# Duplicates
print(df.duplicated().sum())
```

Univariate Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Distribution of rentals
sns.histplot(df['cnt'], bins=30, kde=True)
plt.title("Distribution of Daily Bike Rentals")
plt.show()

# Rentals by season
sns.boxplot(x="season", y="cnt", data=df)
plt.title("Bike Rentals by Season")
plt.show()
```

Bivariate Analysis

```python
# Rentals vs Temperature
sns.scatterplot(x="temp", y="cnt", data=df)
plt.title("Effect of Temperature on Rentals")
plt.show()

# Correlation heatmap
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

Time Series Trends

```python
# Convert date column
df['dteday'] = pd.to_datetime(df['dteday'])

# Rentals over time
plt.figure(figsize=(12,6))
sns.lineplot(x="dteday", y="cnt", data=df)
plt.title("Daily Bike Rentals Over Time")
plt.show()
```

Insights to Report

From this dataset, you’ll likely find:
