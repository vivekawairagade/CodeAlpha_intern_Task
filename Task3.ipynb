Data Story

Objective  
The goal of this analysis was to understand **patterns in daily bike rentals** and identify the key factors (season, weather, temperature) that influence demand.  

Key Observations  
1. **Seasonal Trends**  
   - Rentals are **highest in summer (season = 3)** and **lowest in winter (season = 1)**.  
   - This shows that bike usage is strongly tied to weather conditions.  

2. **Monthly Growth**  
   - Rentals steadily increase from **spring (March–May)** and peak around **July–August**.  
   - Winter months (December–January) show a clear dip in demand.  

3. **Temperature Effect**  
   - Scatterplots reveal a **positive correlation between temperature and rentals**.  
   - Warmer days encourage more biking, while colder days reduce demand.  


 Insights  
- **Commuter Behavior:** Registered users rely on bikes for daily travel, especially during working days.  
- **Seasonality:** Weather and temperature are the strongest drivers of demand.  
- **Weekend Leisure:** Casual rentals rise on Saturdays and Sundays, showing recreational usage.  

 Actionable Takeaways  
- **Business Strategy:** Offer **winter discounts** or promotions to balance seasonal demand.  
- **Operations:** Ensure **extra bikes are available in summer** to meet peak demand.  
- **Marketing:** Target **weekend campaigns** toward casual riders (tourists, leisure users).
